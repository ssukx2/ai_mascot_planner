## 함수화

```python
import requests
from bs4 import BeautifulSoup
```

```python
def dig(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser", from_encoding='cp949')

    return soup
```

```python
def emoji_image(url):
    res = dig(url)
    return res.find("meta", {"name" : "twitter:image"}).get("content")
```

```python
import os
```

```python
file_name = os.path.join(save_dir, 'saved_image.jpg')
```

```python
def save_image(url):
    # 이미지 url 추출
    image_url = emoji_image(url)

    # 이미지 파일 저장
    save_dir = "/Users/seulkim/Desktop/아삭ASAC/Team_DL_Project/emoji"

    def get_unique_filename(file_path):
        base, extension = os.path.splitext(file_path)
        counter = 1
        while os.path.exists(file_path):
            file_path = f"{base} ({counter}){extension}"
            counter += 1
        return file_path
    
    response = requests.get(image_url)
    file_name = os.path.join(save_dir, 'saved_image.jpg')
    if response.status_code == 200:
        unique_file_name = get_unique_filename(file_name)
        with open(unique_file_name, 'wb') as file:
            file.write(response.content)
        print(f"Image saved as '{unique_file_name}'")
    else:
        print("Failed to retrieve the image.")
```

## crop + save

```python
import os
from PIL import Image

save_dir = "/Users/seulkim/Desktop/아삭ASAC/Team_DL_Project/emoji"
crop_dir = os.path.join(save_dir, "cropped")
os.makedirs(crop_dir, exist_ok=True)

def crop_image(image_path, save_dir, crop_dir):
    base_name = os.path.splitext(os.path.basename(image_path))[0]
    img = Image.open(image_path)
    img_width, img_height = img.size
    crop_width = img_width // 3
    crop_height = img_height // 2

    for i in range(2):  # rows
        for j in range(3):  # columns
            left = j * crop_width
            upper = i * crop_height
            right = (j + 1) * crop_width
            lower = (i + 1) * crop_height
            cropped_img = img.crop((left, upper, right, lower))
            cropped_img_name = f"{base_name}_part_{i*3+j+1}.jpg"
            cropped_img.save(os.path.join(crop_dir, cropped_img_name))
            print(f"Saved {cropped_img_name}")
```

```python
for file_name in os.listdir(save_dir):
    file_path = os.path.join(save_dir, file_name)
    if os.path.isfile(file_path) and file_path.endswith(('.jpg', '.jpeg', '.png')):
        crop_image(file_path, save_dir, crop_dir)
```
